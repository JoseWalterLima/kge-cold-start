```python
import pandas as pd
import json
PATH = 'data/'
```

```python
w = pd.read_csv(PATH + 'watchedRel.csv')
w = w.merge(pd.read_csv(PATH + 'releaseRel.csv'), on='movieId', how='left')
w = w.merge(pd.read_csv(PATH + 'movieNode.csv'), on='movieId', how='left')
w
```

```python
# keep only ratings for movies released in 2000
w_2000 = w[(w['releaseDate'] == 2000) | (w['releaseDate'] == 1999)]

# count unique users per movie and keep those with at least 50 users
user_counts_2000 = w_2000.groupby('movieId')['userId'].nunique()
popular_movie_ids_2000 = user_counts_2000[user_counts_2000 >= 50].index

# filtered dataframe: movies from 2000 with >= 50 users
w_popular_2000 = w_2000[w_2000['movieId'].isin(popular_movie_ids_2000)]

w_popular_2000
```

```python
w_popular_2000 = w_popular_2000.drop_duplicates(subset=["movieTitle"])[["movieId", "movieTitle"]]
w_popular_2000
```

```python
records = w_popular_2000.copy()
records['movieId'] = records['movieId'].astype(str)
records = records.to_dict(orient='records')
out_path = 'experiments/test_ids.json'
with open(out_path, 'w', encoding='utf-8') as f:
    json.dump(records, f, ensure_ascii=False, indent=4)
```
