```python
# verificar compatibilidade de epochs
if losses_archA.ndim != 2 or losses_archB.ndim != 2:
    raise RuntimeError("losses_archA e losses_archB devem ter shape (R, epochs)")

E = losses_archA.shape[1]
if losses_archB.shape[1] != E:
    raise RuntimeError("Nmero de epochs diferente entre arquiteturas")

epochs = np.arange(1, E + 1)

meanA = losses_archA.mean(axis=0)
stdA  = losses_archA.std(axis=0)
meanB = losses_archB.mean(axis=0)
stdB  = losses_archB.std(axis=0)

plt.figure(figsize=(8,5))
plt.fill_between(epochs, meanA - stdA, meanA + stdA, alpha=0.2, color='C0')
plt.plot(epochs, meanA, label='Arch A mean', color='C0', linewidth=2)
plt.fill_between(epochs, meanB - stdB, meanB + stdB, alpha=0.2, color='C1')
plt.plot(epochs, meanB, label='Arch B mean', color='C1', linewidth=2)

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training loss: mean  std across runs')
plt.legend()
plt.grid(alpha=0.25)
plt.tight_layout()
# salvar figura se desejar
# plt.savefig("loss_curves_mean_std.png", dpi=300)
plt.show()
```
